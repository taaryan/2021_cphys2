```python
# 학번 : 2016017011
# 학과 : 물리학과
# 이름 : 한승대
# 프로젝트 명 : Gradient Descent
# 프로그램 설명 : gradient descent 방법을 적용하여 함수의 최저점을 구하는 프로그램

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.01)
```

```python
def f(x):
    return 0.5*x**4 - 3*x**3
plt.ylim(-100, 100)
plt.xlim(-5, 7)
plt.plot(x, f(x))

delta = 0.001
a = 6
```

```python
for i in range(10000):
    slope = f(a + delta) - f(a)
    a = a - 0.1 * slope

plt.ylim(-100, 100)
plt.xlim(-5, 7)
plt.plot(x, f(x))
plt.plot(a, f(a), 'o', color='red')
```
