```python
# 학번 : 2016017011
# 학과 : 물리학과
# 이름 : 한승대
# 프로젝트 명 : Diabetes (regression)
# 프로그램 설명 : skelarn을 사용하지 않고 당뇨 데이터를 회귀 분석하는 프로그램

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
diabetes=datasets.load_diabetes()
diabetes.keys()
diabetes.DESCR
```

```python
X_age=diabetes.data[:,0] # 나이
X_bmi=diabetes.data[:,2] # BMI
X_p=diabetes.data[:,3] # 혈압
Y = diabetes.target
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.plot(X_age, Y, '.')
plt.subplot(1,3,2)
plt.plot(X_bmi, Y, '.')
plt.subplot(1,3,3)
plt.plot(X_p, Y, '.')
```

```python
def f(a,b,xi):
    return a*xi+b
def mse(a, b, x, y):
    y_hat=a*x+b
    return np.mean((y_hat-y)**2)
```

```python
eps = 0.1
n= 0
a, b = 1, 200
while(n<30000):
    if(mse(a+eps,b,X_age,Y)>mse(a-eps,b,X_age,Y)):
        a=a-eps
    else:
        a=a+eps
    
    if(mse(a,b+eps,X_age,Y)>mse(a,b-eps,X_age,Y)):
        b=b-eps
    else:
        b=b+eps
    n=n+1
a, b
```

```python
plt.plot(X_age,Y, '.')
plt.plot(X_age,f(a,b,X_age)) # 나이에 대한 regression
```

```python
plt.plot(X_bmi,Y, '.')
plt.plot(X_bmi,f(a,b,X_bmi)) # bmi에 대한 regression
```
