```python
# 학번 : 2016017011
# 학과 : 물리학과
# 이름 : 한승대
# 프로젝트 명 : Diabetes (regression)
# 프로그램 설명 : skelarn을 사용하여 당뇨 데이터를 회귀 분석하는 프로그램

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
```

```python
diabetes=datasets.load_diabetes()
diabetes.keys()
diabetes.DESCR
```

```python
X_age=diabetes.data[:,0] # 나이
X_bmi=diabetes.data[:,2] # BMI
X_p=diabetes.data[:,3] # 혈압
Y = diabetes.target
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.plot(X_age, Y, '.')
plt.subplot(1,3,2)
plt.plot(X_bmi, Y, '.')
plt.subplot(1,3,3)
plt.plot(X_p, Y, '.')
```

```python
X_age = np.array([[i] for i in X_age])
X_bmi = np.array([[i] for i in X_bmi])
X_p = np.array([[i] for i in X_p])
Y = np.array([[i] for i in Y])
```

```python
f=LinearRegression()
f.fit(X_age,Y)
f.fit(X_bmi,Y)
f.fit(X_p,Y)
```

```python
f.coef_, f.intercept_
```

```python
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.plot(X_bmi,Y,'.')
plt.plot(X_bmi,f.predict(X_bmi))
plt.subplot(1,3,2)
plt.plot(X_age, Y, '.')
plt.plot(X_age,f.predict(X_age))
plt.subplot(1,3,3)
plt.plot(X_p, Y, '.')
plt.plot(X_p,f.predict(X_p))
```
