```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
diabetes = datasets.load_diabetes()
diabetes.keys()
diabetes.DESCR
```

```python
X_p = diabetes.data[:, 2]  # 혈압
X_d = diabetes.target  # 당뇨 진행 정도
plt.plot(X_p, X_d, '.')
plt.plot([-0.1], [100], 'r*')
plt.plot([0], [200], 'k*')
plt.plot([0.1], [300], 'b*')
```

```python
temp = []
for (i, j) in zip(X_p, X_d):
    r1 = (i+0.1)**2 + (j-100)**2
    r2 = i**2 + (j-200)**2
    r3 = (i-0.1)**2 + (j-300)**2
    m = min(r1, r2, r3)
    temp.append([r1, r2, r3].index(m))
```

```python
x1 = np.transpose([[i, j] for (t, i, j) in zip(temp, X_p, X_d) if(t == 0)])
x2 = np.transpose([[i, j] for (t, i, j) in zip(temp, X_p, X_d) if(t == 1)])
x3 = np.transpose([[i, j] for (t, i, j) in zip(temp, X_p, X_d) if(t == 2)])
r1 = np.mean(x1[0]), np.mean(x1[1])
r2 = np.mean(x2[0]), np.mean(x2[1])
r3 = np.mean(x3[0]), np.mean(x3[1])
```

```python
plt.scatter(X_p, X_d, c=temp)
plt.plot(r1[0], r1[1], 'r*')
plt.plot(r2[0], r2[1], 'k*')
plt.plot(r3[0], r3[1], 'b*')
```

```python
err1 = np.mean(x1[0] - r1[0]) ** 2 + np.mean(x1[1] - r1[1]) ** 2
err2 = np.mean(x2[0] - r2[0]) ** 2 + np.mean(x2[1] - r2[1]) ** 2
err3 = np.mean(x3[0] - r3[0]) ** 2 + np.mean(x3[1] - r3[1]) ** 2
e_total = (err1 + err2 + err3) * 10 ** 30
```

```python
plt.plot(X_p, X_d, '.')
plt.plot([-0.1], [50], 'r*')
plt.plot([-0.02], [130], 'k*')
plt.plot([0.06], [210], 'b*')
plt.plot([0.14], [290], 'g*')
```

```python
temp_ = []
for (i, j) in zip(X_p, X_d):
    r1_ = (i + 0.1) ** 2 + (j-50) ** 2
    r2_ = (i + 0.02) ** 2 + (j-130) ** 2
    r3_ = (i - 0.06) ** 2 + (j-210) ** 2
    r4_ = (i - 0.14) ** 2 + (j - 290) ** 2
    m = min(r1_, r2_, r3_, r4_)
    temp_.append([r1_, r2_, r3_, r4_].index(m))
```

```python
x1_ = np.transpose([[i, j] for (t, i, j) in zip(temp_, X_p, X_d) if(t == 0)])
x2_ = np.transpose([[i, j] for (t, i, j) in zip(temp_, X_p, X_d) if(t == 1)])
x3_ = np.transpose([[i, j] for (t, i, j) in zip(temp_, X_p, X_d) if(t == 2)])
x4_ = np.transpose([[i, j] for (t, i, j) in zip(temp_, X_p, X_d) if(t == 3)])
r1_ = np.mean(x1_[0]), np.mean(x1_[1])
r2_ = np.mean(x2_[0]), np.mean(x2_[1])
r3_ = np.mean(x3_[0]), np.mean(x3_[1])
r4_ = np.mean(x4_[0]), np.mean(x4_[1])
```

```python
plt.scatter(X_p, X_d, c=temp_)
plt.plot(r1_[0], r1_[1], 'r*')
plt.plot(r2_[0], r2_[1], 'k*')
plt.plot(r3_[0], r3_[1], 'b*')
plt.plot(r4_[0], r4_[1], 'g*')
```

```python
err1_ = np.mean(x1_[0] - r1_[0]) ** 2 + np.mean(x1_[1] - r1_[1]) ** 2
err2_ = np.mean(x2_[0] - r2_[0]) ** 2 + np.mean(x2_[1] - r2_[1]) ** 2
err3_ = np.mean(x3_[0] - r3_[0]) ** 2 + np.mean(x3_[1] - r3_[1]) ** 2
err4_ = np.mean(x4_[0] - r4_[0]) ** 2 + np.mean(x4_[1] - r4_[1]) ** 2
e_total_ = (err1_ + err2_ + err3_ + err4_) * 10 ** 30
```
